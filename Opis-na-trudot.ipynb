# What can Republic od Macedonia learn from Slovenian transport planning?
Станува збор за **оригинален истражувачки труд** во полето на сообраќајот и транспортот, односно што може да учи Република Македонија од Словенските планирања за транспорт. Целта е модернизирање на ***патиштата, автомобилите, аеродромите***.. со цел да стигнат на ниво на европските мерки и на тој начин се повеќе да се доближуваме до **Европската унија**. Транспортот обезбедува многу придобивки за опшеството и економијата од една страна, но од друга страна се негативните ефекти врз здравјето на луѓето и животната средина. За таа цел да имаме ефикасен транспортен систем, мора да се спроведи национална транспортна политика фокусирана од стратешки документи. Република Словенија и Република Македонија во одреден степен се слични во националната политика за транспорт, но од друга страна се разликуваат во социо-економскиот развој. Транспортот обезбедува многу придобивки за опшеството. Тоа му дава на поединецот слобода и независност да патува. Во денешно време планирањето на транспорт има за цел да постигне оддржлив транспортен систем кој е *достапен, безбеден и еколошки*. Целта на овај труд е да се постигне безбедност во сообраќајот и јавниот живот, превоз на патници и да обезбеди сет на предлози за македонските истражувачи кои се вклучени во поставување на национален превоз. Нашето постигнување се заснова на постигнувањата на Република Словенија, која како членка во Европската унија покажа дека е солидна во:
1. мобилноста 
2. пристапноста
3. подобрената услуга за економија
4. безбедност во сообраќајот
5. пониски транспортни трошоци за корисниците. 

Како заклучок ова од никој случај не е лесно и едноставно, но сепак не е недостижно.
